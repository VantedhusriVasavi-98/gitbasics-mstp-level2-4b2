```python
## creating datasets
import pandas as pd
bosten_df=pd.DataFrame(bosten.data,columns=bosten.feature_names)
bosten_df.head()
```

```python
bosten_df["Land price"]=bosten.target
```

```python
bosten_df.head()
```

```python
x=bosten_df.drop("Land price",axis=1)
x
```

```python
x.values
```

```python
y=bosten_df["Land price"]
y.values
```

```python
## Redge Regression=sum of error+sum of sqres magnitude of coefficient(alpha)
from sklearn.linear_model import Ridge
rr=Ridge(alpha=1,solver="svd")
rr
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
rr.fit(x_train,y_train)
```

```python
y_predict=rr.predict(x_train)
y_predict
```

```python
rr.score(x_train,y_train)
```

```python
rr.score(x_test,y_test)
```

```python
rr.coef_
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lrtrain_score=lr.score(x_train,y_train)
print("LR Train score",lrtrain_score)
```

```python
lrtest_score=lr.score(x_test,y_test)
print("LR Testing score",lrtest_score)
```

```python
import matplotlib.pyplot as plt
plt.plot(rr.coef_,marker="",label="Ridge coefficient")
plt.plot(lr.coef_,marker="",label="Linear coeffi")
plt.xlabel("coefficient index")
plt.ylabel("coefficient magnitude")
plt.title("Ridge vs Linear")
plt.legend()
plt.show()
```
