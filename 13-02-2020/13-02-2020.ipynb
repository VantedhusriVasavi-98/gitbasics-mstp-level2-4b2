## Decision Tree

```python
A desicion tree is a supervised Machine learning algorithm looks like a tree.Wherein each node
represents a predictor variable the link between the nodes represents a Decision and each leaf
node represents an outcome 
```

## Entropy

DecisionTree can split the data

```python
import pandas as pd
d=[[3,4,'dog'],[90,90,'cat'],[22,22,'cat'],[23,23,'dog']]
df=pd.DataFrame(d,columns=['height','weight','pet'])
df
```

```python
#split data
x=df.drop('pet',axis=1)
x
```

```python
y=df['pet']
y
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf= DecisionTreeClassifier()
```

```python
clf.fit(x,y)
```

```python
clf.predict([[22,22]])
```

```python
#graphviz-->plot
#pydotplus-->dot file into image
```

```python
from sklearn.externals.six import StringIO
from sklearn.tree import export_graphviz
import pydotplus
from IPython.display import Image
d1=StringIO()
export_graphviz(clf,out_file=d1,
               filled=True,rounded=True,
               feature_names=x.columns,
               class_names=['cat','dog'])
```

```python
graph=pydotplus.graph_from_dot_data(d1.getvalue())
Image(graph.create_png())
```
