## Matplotlib ticks
### Ticks is values within a coordinates

```python
import matplotlib.pyplot as p1 
x=list(range(5,51,5))
print(x)
y=list(range(1,11))
print(y)
p1.plot(x,y)
p1.title("A Graph a withexample Ticks")
p1.xlabel('Natural Numbers')
p1.ylabel('value with ++5 up to 50')
p1.show()
```

## Representing a sine wave
* A=sine(wt)A=amplitude w frequency t =time

```python
import numpy as np
import matplotlib.pyplot as p1
Time=np.arange(1,10,0.1)
Amplitude=np.sin(Time)
p1.grid()
p1.plot(Time,Amplitude)
p1.title("sine wavw A=sin(wt)")
p1.xlabel('Time')
p1.ylabel('Amplitude')
p1.show()
```

```python
import pandas as pd
df=pd.read_csv('income.csv')
df
```

```python
import matplotlib.pyplot as p1
years=df.columns[2::]
Alaska_income=df.iloc[1,2:]
p1.plot(years,Alaska_income)
p1.title("Income of alaska")
p1.xlabel("years")
p1.ylabel("Income of Alaska")
p1.savefig("Alaska income")
```

```python
# plot of all state income from 2005-20013
import matplotlib.pyplot as p1
import pandas as pd
years=df.columns[-5::]
income=df.iloc[:,2:]
p1.plot(years,income)
p1.title("Income of all states")
p1.xlabel("years")
p1.ylabel("Income")
p1.show()
p1.savefig("Income of all ststes.jpg")

```
